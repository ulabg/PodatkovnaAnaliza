## 1.TABELA

### List of countries and dependencies and their capitals in native languages

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re

url = 'https://kt.ijs.si/~ljupco/lectures/papvp-2425/podatkovni-viri/List%20of%20countries%20and%20dependencies%20and%20their%20capitals%20in%20native%20languages%20-%20Wikipedia.html'

povezava = requests.get(url)
spletna_stran = BeautifulSoup(povezava.content, 'html.parser') #prenos HTML vsebine s spletne strani

tabele = spletna_stran.find_all('table', {'class': 'wikitable'}) #iskanje vseh tabel na strani z razredom 'wikitable'

seznami_podatkov = [pd.read_html(str(tabela))[0] for tabela in tabele] # pretvorba vsake tabele v Pandas DataFramež

zdruzena_tabela = pd.concat(seznami_podatkov, ignore_index=True)
print(zdruzena_tabela)


nova_tabela = zdruzena_tabela[['Country (exonym)', 'Capital (exonym)']].copy()
nova_tabela.columns = ['drzava', 'ime']

# Funkcija za čiščenje držav
def ocisti_drzavo(drzava):
    if isinstance(drzava, str):  # Preverimo, če je vrednost niz (string)
        drzava = re.sub(r'\[.*?\]', '', drzava)  # Odstrani oglati oklepaji
        drzava = re.sub(r'\(.*?\)', '', drzava)  # Odstrani okrogli oklepaji
        return drzava.strip()  # Odstrani odvečne presledke
    return drzava  # Če je NaN, pustimo nespremenjeno

# Funkcija za čiščenje glavnih mest
def ocisti_mesto(mesto):
    if isinstance(mesto, str):  # Preverimo, če je niz (string)
        mesto = re.split(r'\(|\[', mesto)[0]  # Odstrani vse za prvim oklepajem (okrogel ali oglati)
        return mesto.strip()  # Odstrani odvečne presledke
    return mesto
    
# Uporaba funkcij za čiščenje
nova_tabela['drzava'] = nova_tabela['drzava'].apply(ocisti_drzavo)
nova_tabela['ime'] = nova_tabela['ime'].apply(ocisti_mesto)
print(nova_tabela)

# v csv 
glavna_mesta = "glavna_mesta.csv"
nova_tabela.to_csv(glavna_mesta, index=False)
print(f"Podatki so shranjeni v datoteko: {glavna_mesta}")
```

## 2.TABELA

### Population Growth Rates in Urban areas and Capital cities (undata)

```python
import pandas as pd

vhod = "https://kt.ijs.si/~ljupco/lectures/papvp-2425/podatkovni-viri/SYB61_253_Population%20Growth%20Rates%20in%20Urban%20areas%20and%20Capital%20cities.csv"

# Branje CSV datoteke
podatki = pd.read_csv(vhod)

#odstranimo header vrstico
tabela = podatki.iloc[1:]

#nastavimo stolpce
tabela.columns = podatki.iloc[0]

tabela = tabela.reset_index(drop=True)

tabela.rename(columns={
    'Year': 'leto',
    'Series': 'spremenljivka',
    'Value': 'vrednost'},
    inplace=True)

tabela = tabela.rename(columns={tabela.columns[1]: 'enota'})

koncna_tabela = tabela[tabela['spremenljivka'] != 'Rural population (percent growth rate per annum)']
koncna_tabela = koncna_tabela[koncna_tabela['spremenljivka'] != 'Urban population (percent growth rate per annum)']
koncna_tabela = koncna_tabela[koncna_tabela['spremenljivka'] != 'Capital city population (as a percentage of total urban population)']

populacija_koncna = pd.DataFrame(koncna_tabela, columns=['enota', 'leto','spremenljivka','vrednost'])
populacija_koncna = populacija_koncna.iloc[4:]
populacija_koncna = populacija_koncna.reset_index(drop=True)
populacija_koncna['vrednost'] = populacija_koncna['vrednost'].fillna(0)
populacija_koncna.to_csv('populacija.csv', index=False)
```

## 3.NALOGA

### List of Academy Award–winning films

```python
import pandas as pd

# URL do Wikipedije
url = "https://en.wikipedia.org/wiki/List_of_Academy_Award%E2%80%93winning_films"

# Branje tabel iz URL-ja
tabele = pd.read_html(url)

# Preverjanje števila tabel
print(f"Najdene tabele: {len(tabele)}")

# Izbor želene tabele (privzeto je prva tabela prava)
tabela_nagrade = tabele[0]

# Prikaz prvih vrstic za pregled
print(tabela_nagrade.head())

# Preimenovanje stolpcev za boljšo berljivost
tabela_nagrade.columns = [
    "naslov",
    "leto",
    "oskarji_osvojeni",
    "oskarji_nominacije"
]

# Odstranitev morebitnih praznih vrstic
tabela_nagrade = tabela_nagrade.dropna()

# Pretvorba leta v prvo število pred znakom '/'
tabela_nagrade["leto"] = tabela_nagrade["leto"].str.split('/').str[0]

# Čiščenje podatkov v stolpcih "oskarji_osvojeni" in "oskarji_nominacije"
tabela_nagrade["oskarji_osvojeni"] = tabela_nagrade["oskarji_osvojeni"].str.extract(r"(\d+)").astype(int)
tabela_nagrade["oskarji_nominacije"] = tabela_nagrade["oskarji_nominacije"].str.extract(r"(\d+)").astype(int)

# Priprava tabele z razdeljenimi vrsticami
razsirjene_vrstice = []
for _, vrstica in tabela_nagrade.iterrows():
    osvojeni = vrstica["oskarji_osvojeni"]
    nominirani = vrstica["oskarji_nominacije"]
    neosvojeni = nominirani - osvojeni
    # Dodaj vrstico za FALSE (kolikokrat ni osvojil)
    razsirjene_vrstice.append([vrstica["naslov"], vrstica["leto"], "FALSE", neosvojeni])
    # Dodaj vrstico za TRUE (kolikokrat je osvojil)
    razsirjene_vrstice.append([vrstica["naslov"], vrstica["leto"], "TRUE", osvojeni])

# Ustvarjanje nove tabele
koncna_tabela = pd.DataFrame(razsirjene_vrstice, columns=["naslov", "leto", "oskar", "stevilo"])

# Odstranitev filmov, kjer so vse nominacije in zmage enake 0
koncna_tabela = koncna_tabela.groupby("naslov").filter(lambda x: not (x["stevilo"].sum() == 0))

# Izvoz v CSV
koncna_pot_csv = "filmi.csv"
koncna_tabela.to_csv(koncna_pot_csv, index=False)

print(f"Tabela je bila uspešno shranjena kot {koncna_pot_csv}")
```

## 4.NALOGA

### List of Academy Award–winning foreign-language films

```python
import pandas as pd

# URL strani z Wikipedije
url = 'https://kt.ijs.si/~ljupco/lectures/papvp-2425/podatkovni-viri/List%20of%20Academy%20Award%e2%80%93winning%20foreign-language%20films%20-%20Wikipedia.html'

# Preberite vse tabele s strani
tabele = pd.read_html(url)

# Pridobimo prvo tabelo (predpostavljamo, da je to ta prava)
tabela = tabele[0]

# Prilagodimo imena stolpcev glede na strukturo tabele
tabela.columns = ['leto', 'naslov_nom', 'naslov_orig', 'kategorija', 'opombe', 'drzava', 'jezik']

# Tabela vseh kategorij za boljšo organizacijo kategorij
seznam_kategorij = [
    "Actor in a Leading Role",
    "Actress in a Leading Role",
    "Art Direction",
    "Cinematography",
    "Costume Design",
    "Documentary Feature",
    "Film Editing",
    "Makeup",
    "Music",
    "Original Screenplay",
    "Picture",
    "Production Design",
    "Sound Editing",
    "Visual Effects",
    "Writing",
    "Animated Feature"
]

# Funkcija za razširitev vrstic za kategorije, jezike in države
def razsiri_vrstice(vrstica):
    leto = str(vrstica['leto']).split('(')[0].strip()
    zaporedna_podelitev = ''
    if '(' in str(vrstica['leto']):
        zaporedna_podelitev = str(vrstica['leto']).split('(')[1].replace('th)', '').replace('st)', '').replace('nd)', '').replace('rd)', '').strip()

    naslov_nom = pd.Series(str(vrstica['naslov_nom'])).str.replace(r'\[.*?\]', '', regex=True).str.replace(r'\(.*?\)', '', regex=True).iloc[0].strip()
    naslov_orig = pd.Series(str(vrstica['naslov_orig'])).str.replace(r'\[.*?\]', '', regex=True).str.replace(r'\(.*?\)', '', regex=True).iloc[0].strip()

    surove_kategorije = str(vrstica['kategorija'])
    ociscene_kategorije = pd.Series(surove_kategorije).str.replace(r'\(.*?\)', '', regex=True).iloc[0]
    razdeljene_kategorije = ociscene_kategorije.split(',')

    # Popravi kategorije glede na glavni seznam
    popravljene_kategorije = []
    temp = []
    for kat in razdeljene_kategorije:
        kat = kat.strip()
        if kat in seznam_kategorij:
            popravljene_kategorije.append(kat)
        else:
            temp.append(kat)
            polna_kategorija = " ".join(temp)
            if polna_kategorija in seznam_kategorij:
                popravljene_kategorije.append(polna_kategorija)
                temp = []

    jeziki = [jez.strip() for jez in str(vrstica['jezik']).split(',') if jez.strip()]
    drzave = [drz.strip() for drz in str(vrstica['drzava']).split(',') if drz.strip()]

    if not drzave:  # Če ni držav, dodaj "Neznano"
        drzave = ["Neznano"]

    vrstice = []
    for kategorija in popravljene_kategorije:
        for drzava in drzave:
            for jezik in jeziki:
                vrstice.append({
                    'naslov_nom': naslov_nom,
                    'naslov_orig': naslov_orig,
                    'leto': leto,
                    'zaporedna_podelitev': zaporedna_podelitev,
                    'kategorija': kategorija,
                    'drzava': drzava,
                    'jezik': jezik
                })
    return vrstice

# Razširimo vrstice
podrobne_vrstice = []
for _, vrstica in tabela.iterrows():
    podrobne_vrstice.extend(razsiri_vrstice(vrstica))

# Preverimo število vrstic glede na kategorije, države in jezike
razsirjena_tabela = pd.DataFrame(podrobne_vrstice)

# Shranimo tabelo v CSV datoteko
razsirjena_tabela.to_csv('tuji_filmi.csv', index=False, encoding='utf-8-sig')

print("Tabela je bila shranjena v 'tuji_filmi.csv'")
```

## 5.NALOGA

### List of actors with Academy Award nominations

```python
import pandas as pd
import re

# Korak 1: Branje HTML tabele neposredno iz URL-ja
url = 'https://kt.ijs.si/~ljupco/lectures/papvp-2425/podatkovni-viri/List%20of%20actors%20with%20Academy%20Award%20nominations%20-%20Wikipedia.html'
tabele = pd.read_html(url, header=0)  # Preberemo tabelo in uporabimo prvo vrstico kot glavo

# Predpostavimo, da je druga tabela tista, ki nas zanima
podatki = tabele[1]

# Preverimo, kakšna imena stolpcev imamo
print("Imena stolpcev v tabeli:")
print(podatki.columns)

# Uporabimo pravilna imena stolpcev, kot jih prebere pandas
stolpec_igralec = podatki.columns[0]  # Prvi stolpec (igralci)
stolpec_rojstvo = podatki.columns[2]  # Stolpec z letom rojstva
stolpec_smrti = podatki.columns[3]  # Stolpec z letom smrti
stolpec_nominacije = podatki.columns[5]  # Stolpec z nominacijami
stolpec_zmage = podatki.columns[6]  # Stolpec z zmagami
stolpec_glavna_stranska = podatki.columns[7]  # Stolpec z razdelitvijo vlog
stolpec_film = podatki.columns[8]  # Stolpec s filmi

ociščeni_podatki = []

# Funkcija za čiščenje številskih vrednosti (odstrani opombe in preveri veljavnost)
def ocisti_stevilsko(vrednost):
    if pd.isnull(vrednost):
        return 0
    vrednost = str(vrednost)
    vrednost = re.sub(r'\[.*?\]', '', vrednost)  # Odstranimo vse v oglatih oklepajih
    vrednost = vrednost.strip()  # Odstranimo morebitne presledke
    try:
        return int(vrednost) if vrednost.isdigit() else 0
    except ValueError:
        return 0

# Korak 2: Obdelava podatkov iz stolpcev
for _, vrstica in podatki.iterrows():
    # Pridobimo podatke iz vsake vrstice
    igralec = vrstica[stolpec_igralec]
    leto_rojstva = vrstica[stolpec_rojstvo]
    leto_smrti = vrstica[stolpec_smrti] if vrstica[stolpec_smrti] != "~" else None
    nominacije = ocisti_stevilsko(vrstica[stolpec_nominacije])  # Očistimo nominacije
    zmage = ocisti_stevilsko(vrstica[stolpec_zmage])  # Očistimo zmage
    podrobnosti_vlog = vrstica[stolpec_glavna_stranska]
    film = vrstica[stolpec_film]

    # Vedno razdelimo nominacije na podlagi Glavna in Stranska podrobnosti
    podrobnosti_razdeljene = podrobnosti_vlog.split(':')
    glavne_nominacije = (
        int(podrobnosti_razdeljene[0].strip('L')) if len(podrobnosti_razdeljene) > 0 and podrobnosti_razdeljene[0].strip('L').isdigit() else 0
    )
    stranske_nominacije = (
        int(podrobnosti_razdeljene[1].strip('S')) if len(podrobnosti_razdeljene) > 1 and podrobnosti_razdeljene[1].strip('S').isdigit() else 0
    )

    # Če nominacije ne ustrezajo razdelitvi, jih ročno razdelimo
    if glavne_nominacije + stranske_nominacije != nominacije:
        glavne_nominacije = nominacije - stranske_nominacije
        stranske_nominacije = nominacije - glavne_nominacije

    # Pravilna obdelava zmag
    glavne_zmage = min(zmage, glavne_nominacije)  # Zmage pri glavnih vlogah
    stranske_zmage = zmage - glavne_zmage  # Preostale zmage so pri stranskih

    # Pravilna razporeditev nominacij in zmag
    for glavna in ['Glavna', 'Stranska']:
        je_glavna = (glavna == 'Glavna')
        for status_zmage in [False, True]:  # Obrnemo vrstni red, da je najprej False, nato True
            ociščeni_podatki.append({
                'Ime': igralec,
                'Leto_rojstva': leto_rojstva,
                'Leto_smrti': leto_smrti,
                'Nominacija': glavna,
                'Oskar': status_zmage,
                'Število': (glavne_zmage if je_glavna and status_zmage else
                            (glavne_nominacije - glavne_zmage) if je_glavna and not status_zmage else
                            stranske_zmage if not je_glavna and status_zmage else
                            (stranske_nominacije - stranske_zmage) if not je_glavna and not status_zmage else 0),
                'Prv_film': film
            })

# Korak 3: Ustvarjanje tabele
ociščena_tabela = pd.DataFrame(ociščeni_podatki)

# Korak 4: Shranjevanje kot CSV
ociščena_tabela.to_csv('igralci.csv', index=False)
print("Tabela je bila shranjena kot 'igralci.csv'.")
```
