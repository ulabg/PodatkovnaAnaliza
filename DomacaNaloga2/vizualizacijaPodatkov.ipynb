```python
import matplotlib.pyplot as plt

# Graf a: ≈†tevilo podeljenih oskarjev
def graf_podeljeni_oskarji(leta, stevilo_oskarjev):
    plt.figure(figsize=(10, 6))
    plt.plot(leta, stevilo_oskarjev, marker='o', color='red', linestyle='-', linewidth=2)
    plt.title("Oskarji\n≈†tevilo podeljenih oskarjev med leti 1927 in 2022", fontsize=14)
    plt.xlabel("Leto", fontsize=12)
    plt.ylabel("≈†tevilo podeljenih oskarjev", fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.savefig("graf3a.pdf")
    plt.show()

# Podatki za graf a (dopolnite z va≈°imi podatki)
leta = [1927, 1928, 1929, 1930, 1931]  # Dodajte vsa leta
stevilo_oskarjev = [12, 15, 20, 25, 22]  # Dodajte ≈°tevilo podeljenih oskarjev

graf_podeljeni_oskarji(leta, stevilo_oskarjev)

# Graf b: Odstotek prebivalstva v glavnem mestu
def graf_prebivalstvo_glavno_mesto(drzave, odstotek_glavno_mesto, odstotek_izven):
    bar_width = 0.5
    plt.figure(figsize=(10, 6))
    bar_positions = np.arange(len(drzave))
    plt.bar(bar_positions, odstotek_glavno_mesto, color='gray', label="glavno mesto", width=bar_width)
    plt.bar(bar_positions, odstotek_izven, color='green', bottom=odstotek_glavno_mesto, label="izven", width=bar_width)
    plt.xticks(bar_positions, drzave, rotation=45, fontsize=10)
    plt.yticks(fontsize=10)
    plt.title("Odstotek prebivalstva v glavnem mestu leta 2015", fontsize=14)
    plt.ylabel("Odstotek", fontsize=12)
    for i, (g, i_) in enumerate(zip(odstotek_glavno_mesto, odstotek_izven)):  # Dodamo odstotke na stolpce
        plt.text(i, g / 2, f"{g}%", ha='center', fontsize=10, color='black')
        plt.text(i, g + i_ / 2, f"{i_}%", ha='center', fontsize=10, color='white')
    plt.legend(fontsize=10)
    plt.tight_layout()
    plt.savefig("graf3b.pdf")
    plt.show()

# Podatki za graf b (dopolnite z va≈°imi podatki)
drzave = ["Portoriko", "Paragvaj", "Panama", "Peru", "Portugalska"]
odstotek_glavno_mesto = [67.1, 44.7, 42.2, 31.3, 27.7]  # Odstotki prebivalstva v glavnem mestu
odstotek_izven = [32.9, 55.3, 57.8, 68.7, 72.3]  # Odstotki prebivalstva izven glavnega mesta

graf_prebivalstvo_glavno_mesto(drzave, odstotek_glavno_mesto, odstotek_izven)
```
